matplotlib.pyplot

Se utiliza para crear gráficos y visualizar datos.

```python
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100) # Generar 100 puntos de -10 a 10
y = x**2 # Funcion Cuadratica

plt.plot(x, y)
plt.title("Grafco de f(x) = x^2")
plt.xlabel("X")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
